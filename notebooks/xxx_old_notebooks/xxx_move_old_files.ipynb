```python
import glob
import shutil
import os
```

```python


source_directory = "../../data/polynya_size/"
# destination_directory = "/Users/mackenziejewell/Desktop/to_delete/"
destination_directory = "../../data/polynya_size/previous_versions/"


file_extension = "*v7.6*"  # Example: move all .txt files


all_files = glob.glob(os.path.join(source_directory, '*'), recursive=False)
files_to_keep = glob.glob(os.path.join(source_directory, file_extension), recursive=False)
files_to_move = [file for file in all_files if file not in files_to_keep]

print(files_to_keep[:3])
print()
print(files_to_move[:3])
```

```python
# Move files one by one
for file_path in files_to_move:
    try:
        filename = os.path.basename(file_path)
        destination_path = os.path.join(destination_directory, filename)
        shutil.move(file_path, destination_path)
        print(f"Moved: {file_path} -> {destination_path}")
    except Exception as e:
        print(f"Error moving {file_path}: {e}")
```
