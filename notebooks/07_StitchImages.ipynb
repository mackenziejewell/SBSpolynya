```python
from PIL import Image
```

```python
# Open both images
img1 = Image.open('../figures/results/pease_model/timeseries_v2.png')
img2 = Image.open('../figures/results/pease_model/map_panels.png')
```

```python
# Create a new image with width = img1 + img2, and height = max of both
# new_width = img1.width + img2.width
# new_height = max(img1.height, img2.height)


scale_factor = 1.25
new_size = (int(img1.width * scale_factor), int(img1.height * scale_factor))
img1 = img1.resize(new_size, Image.Resampling.LANCZOS)


new_height = img2.height# + img2.height
new_width = img1.width + img2.width

# Create a new blank image with the combined size
new_img = Image.new('RGB', (new_width, new_height), color=(255, 255, 255))

# Paste both images into the new image
new_img.paste(img1, (0, 0))
new_img.paste(img2, (img1.width, 0))
```
